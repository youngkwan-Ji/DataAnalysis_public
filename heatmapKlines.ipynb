```python
import pandas as pd
klines = pd.read_csv('resource/dataset/binance/klines-BTCUSDT-1m_1509.csv')
klines
```

```python
corrKlines = klines.corr()
corrKlines
```

#Dataset 상관관계 분석

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))

cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corrKlines, cmap=cmap, annot=True, square=True, linewidths=.5, cbar_kws={'shrink': .5})
```
