```python
import pandas as pd
study = pd.read_csv('resource/dataset/StudentsStudyHour/Student Study Hour V2.csv')
# study = pd.read_csv('resource/dataset/StudentsStudyHour/Student Study Hour V3.csv')
study.head()

x_study = study[['Hours']]
y_study = study[['Scores']]
# print(x_study.shape, y_study.shape)
# print(x_study.head(2))
# print(y_study.head(2))

x_study['Hours'][0]
# y_study
```

```python
import tensorflow as tf
import numpy as np
# x_train = [1, 2, 3]
# y_train = [1, 2, 3]
x_train = x_study
y_train = y_study

W = tf.Variable(1, name="weight")
b = tf.Variable(1, name="bias")

hypothesis = x_train['Hours']*W+b

cost = tf.reduce_mean(tf.square(hypothesis - y_train['Scores']))
sgd = tf.keras.optimizers.SGD(learning_rate=0.01)

testM = tf.keras.models.Sequential()

testM.add(tf.keras.layers.Dense(1, input_dim = 1,activation='linear'))
# model.compile(loss="mse",)
testM.compile(loss='mse',optimizer=sgd,metrics=['accuracy'])

history = testM.fit(x_train,y_train,epochs=15000,verbose=0,batch_size=1, shuffle=False)

# print(model.predict(np.array([5])))
```
